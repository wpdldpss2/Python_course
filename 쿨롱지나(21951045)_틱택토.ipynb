```python
#틱택토

print('-'*15)
print('  TIC-TAC-TOE  ')
print('-'*15)
print()

table = [' 'for a in range(9)]

def show_table(t, winner=None):
  print(' ' + t[0] + ' ㅣ ' + t[1] + ' ㅣ ' + t[2] +  ' ')
  print('---  ---  ---')
  print(' ' + t[3] + ' ㅣ ' + t[4] + ' ㅣ ' + t[5] +  ' ')
  print('---  ---  ---')
  print(' ' + t[6] + ' ㅣ ' + t[7] + ' ㅣ ' + t[8] +  ' ')
  print()
  if winner:
    print('{} 승리!\n' .format(winner))

def tictactoe():
  player = 'X'
  round = 0

  while True:
    show_table(table)
    print('\n {}선수의 차례. 1에서9까지의 숫자를 골라주세요\n' .format(player)) 
    move = int(input()) - 1 

    if table[move] == ' ':
      table[move] = player
      round += 1
    else:
      print('비어있는 칸으로 골라주세요.\n')
      continue

    if table[0] == table[1] == table [2] != ' ' \
    or table[3] == table[4] == table [5] != ' ' \
    or table[6] == table[7] == table [8] != ' ' \
    or table[0] == table[4] == table [8] != ' ' \
    or table[0] == table[3] == table [6] != ' ' \
    or table[1] == table[4] == table [7] != ' ' \
    or table[2] == table[5] == table [8] != ' ' \
    or table[2] == table[4] == table [6] != ' ' :
       show_table(table, player)
       break
    if round == 9:
      print('무승부!\n')
      break

     
    if player == 'X':
       player = 'O'
    else:
       player = 'X' 

tictactoe()
```
